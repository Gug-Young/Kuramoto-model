```python
from scipy.special import erf

# Example usage
x = np.linspace(-3, 3, 100)
y = erf(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('erf(x)')
plt.title('Error Function')
plt.show()
```

```python
def K_c(m,sigma=1):
    A = 2*np.sqrt(2/np.pi)
    B = 1 - np.exp(1/(2*m**2*sigma**2)) * (1-erf(1/(np.sqrt(2)*m*sigma)))
    return A/B*sigma
```

```python
ms = np.linspace(0,6,100)
plt.plot(ms,K_c(ms))
```
