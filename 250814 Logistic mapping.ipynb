```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm


def logistic_map(x,r):
    return r*x*(1-x)

x = 0.1
r = 3.564407388313484
x_arr = np.zeros(1000)
x_arr[0] = x

for i in range(1,1000):
    x = logistic_map(x,r)
    x_arr[i] = x

plt.plot(x_arr)
plt.show()
np.unique(x_arr[-1000:])
```

# Draw Logistic map

```python
xs  = np.linspace(0,1,1000)
plt.plot(xs,r*xs*(1-xs))
plt.xlim(0,1)
plt.ylim(0,1)
plt.plot(xs,xs)
plt.gca().set_aspect('equal')

# for i in range(0,1000,4):
plt.plot(x_arr[1:],x_arr[:-1])
```
